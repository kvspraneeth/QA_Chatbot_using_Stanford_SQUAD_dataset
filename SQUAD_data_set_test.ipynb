```python
from transformers import pipeline
import gradio as gr
import torch
from transformers import BertForQuestionAnswering, BertTokenizer
```

```python
model_name = "deepset/bert-base-cased-squad2"
model = BertForQuestionAnswering.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
def answer_question(question, context):
    try:
        # Tokenize input
        inputs = tokenizer.encode_plus(question, context, return_tensors="pt")
        # Get answer
        answer_start_scores, answer_end_scores = model(**inputs)
        answer_start = torch.argmax(answer_start_scores)
        answer_end = torch.argmax(answer_end_scores) + 1
        answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))
        return answer
    except Exception as e:
        return str(e)
```

```python
def chatbot(question):
    try:
        answer = answer_question(question)
        return answer
    except:
        return "Error"
```

```python
image = "https://img.freepik.com/premium-photo/robot-with-microphone-his-head-is-black-background_758669-47.jpg"
```

```python
gr.Interface(fn=chatbot,
             inputs="text",
             outputs="text",
             title="SQuAD Chatbot",
             description="Ask any question and get an answer.",
             examples=[["Who is the president of the United States?"]],
             theme="compact",
             allow_flagging=False,
             ).launch()
```
